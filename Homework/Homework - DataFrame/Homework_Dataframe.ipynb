```python
data_paid_transaction["Order Number"] = data_paid_transaction["Order Number"].astype(str)
data_seller["Order Number"] = data_seller["Order Number"].astype(str)
```

```python
data_paid_transaction["Order Number"] = data_paid_transaction["Order Number"].str.replace(r'\D', '', regex=True)
data_seller["Order Number"] = data_seller["Order Number"].str.replace(r'\D', '', regex=True)
```

```python
merged_data = data_paid_transaction.merge(data_seller, left_on="Order Number", right_on="Order Number", how="inner")
```

```python
merged_data
```

# **No 1**

```python
# Menggabungkan kolom "First Name" dan "Last Name" dengan format "title case" menjadi kolom "Full Name"
# str.title() digunakan untuk membuat "title case,"
title_case = merged_data["First Name"].str.title() + " " + merged_data["Last Name"].str.title()

# Mendapatkan indeks kolom "Last Name" dengan cara
# menggunakan get_loc untuk mencari indeks (posisi) dari elemen Last Name
last_name_index = merged_data.columns.get_loc("Last Name")

# Sisipkan kolom "Full Name" setelah kolom "Last Name" menggunakan + 1
merged_data.insert(last_name_index + 1, "Full Name", title_case)
```

```python
merged_data
```

# **No 2**

```python
# Menghitung Gross Merchandise Value (GMV) dengan operasi aritmatika
gmv_hitung = merged_data["Transaction Amount"].astype(int) - merged_data["Seller Discount"].astype(int) + merged_data["Delivery Fee"].astype(int)
#merged_data["GMV"] = merged_data["Transaction Amount"].astype(float) - merged_data["Seller Discount"].astype(float) + merged_data["Delivery Fee"].astype(float)

# Mendapatkan indeks kolom "Seller Discount" dengan cara
# menggunakan get_loc untuk mencari indeks (posisi) dari elemen Seller Discount
seller_discount_index = merged_data.columns.get_loc("Seller Discount")

# Sisipkan kolom "GMV" setelah kolom "Seller Discount" menggunakan + 1
merged_data.insert(seller_discount_index + 1, "GMV", gmv_hitung)
```

```python
merged_data
```

# **No 3**

```python
# Buat dictionary untuk menggabungkan kategori menjadi kelompok yang sesuai
kategori_group = {
    'Home/ Food': 'Group 1',
    'Sports/ Hobi/ Otomotif': 'Group 1',
    'Fashion': 'Group 1',
    'Beauty/ Health': 'Group 1',
    'Gadget/ Komputer': 'Group 2',
    'Elektronik': 'Group 2',
    'Service/ Mokado': 'Group 2',
    'Babies/ Kids': 'Group 2'
}

# Menerapkan pengelompokkan dan di tampung pada variabel "Group" dengan cara menggabungkan kolom Meta Category
# menjadi 2 Group berdasarkan kategori_group yang telah ditentukan
Group = merged_data['Meta Category'].map(kategori_group)
```

```python
Group
```

```python
# Konversi kolom "Paid Date" menjadi tipe datetime  dengan format 'YYYYMMDD' menggunakan pd.to_datetime()
# errors='coerce' untuk menangani jika ada nilai yang tidak sesuai dengan format
# maka kolom "Paid Date" akan diisi dengan nilai NaN
# dan nantinya akan digunakan untuk nomor 4 & 5
merged_data['Paid Date'] = pd.to_datetime(merged_data['Paid Date'], format='%Y%m%d', errors='coerce')

# Ekstrak bulan dari kolom "Paid Date" dan simpan dalam variabel "Month" menggunakan metode .dt.to_period('M').
# untuk mempermudah agregasi data berdasarkan bulan.
Month = merged_data['Paid Date'].dt.to_period('M')

# Buat pivot table untuk jumlah total GMV per-bulan untuk setiap grup.
pivot_table_no3 = merged_data.pivot_table(values='GMV', index=Month, columns=Group, aggfunc='sum')
```

```python
pivot_table_no3
```

```python
df_pivot = pd.DataFrame(pivot_table_no3)
df_pivot
```

# **No 4**

```python
# Filter data untuk bulan Agustus 2017
data_agustus_2017 = merged_data[merged_data['Paid Date'].dt.strftime('%Y-%m') == '2017-08']
data_agustus_2017
```

```python
# Hitung total GMV per seller dengan menggunakan groupby ('Seller') untuk mengelompokkan data
# berdasarkan seller kemudian menghitung total GMV dengan menggunakan metode .sum()
seller_gmv = data_agustus_2017.groupby('Seller')['GMV'].sum().reset_index()
seller_gmv
```

```python
# Mencari seller dengan GMV tertinggi
seller_gmv_tertinggi = seller_gmv[seller_gmv['GMV'] == seller_gmv['GMV'].max()]

# Cetak seller dengan GMV tertinggi di bulan Agustus 2017
print("Seller dengan GMV tertinggi di bulan Agustus 2017:")
print(seller_gmv_tertinggi)
```

```python
seller_gmv_tertinggi
```

# **No 5**

```python
# Filter data untuk bulan September 2017 dengan kategori "Fashion"
data_fashion_september_2017 = merged_data[(merged_data['Paid Date'].dt.strftime('%Y-%m') == '2017-09') & (merged_data['Meta Category'] == 'Fashion')]
```

```python
data_fashion_september_2017
```

```python
# Hitung jumlah transaksi (count) per seller
seller_transaksi_count = data_fashion_september_2017.groupby('Seller')['Order Number'].count().reset_index()
seller_transaksi_count
```

```python
# Temukan seller dengan jumlah transaksi (count) terbanyak
seller_transaksi_terbanyak = seller_transaksi_count[seller_transaksi_count['Order Number'] == seller_transaksi_count['Order Number'].max()]

# Cetak seller dengan jumlah transaksi (count) terbanyak di bulan September 2017 kategori "Fashion"
print("Seller dengan jumlah transaksi (count) terbanyak di bulan September 2017 kategori 'Fashion':")
print(seller_transaksi_terbanyak)
```
