# **NO 2**

```python
from scipy import stats

# Pisahkan data sebelum kampanye, 1 bulan setelah kampanye, dan 2 bulan setelah kampanye
sebelum_kampanye = df['pre_usage']
satu_bulan_pasca_kampanye = df['Post_usage_1month']
dua_bulan_pasca_kampanye = df['post_usage_2ndmonth']

# Uji-t antara sebelum kampanye dan 1 bulan setelah kampanye
t_statistic_1bulan, p_value_1bulan = stats.ttest_rel(sebelum_kampanye, satu_bulan_pasca_kampanye)

# Uji-t antara sebelum kampanye dan 2 bulan setelah kampanye
t_statistic_2bulan, p_value_2bulan = stats.ttest_rel(sebelum_kampanye, dua_bulan_pasca_kampanye)

alpha = 0.05  # Tingkat signifikansi dan sesuai petunjuk

# Hipotesis
H0_1bulan = "Tidak ada perbedaan yang signifikan antara rata-rata penggunaan kartu kredit sebelum kampanye dan 1 bulan setelah kampanye."
H1_1bulan = "Terdapat perbedaan yang signifikan antara rata-rata penggunaan kartu kredit sebelum kampanye dan 1 bulan setelah kampanye."

H0_2bulan = "Tidak ada perbedaan yang signifikan antara rata-rata penggunaan kartu kredit sebelum kampanye dan 2 bulan setelah kampanye."
H1_2bulan = "Terdapat perbedaan yang signifikan antara rata-rata penggunaan kartu kredit sebelum kampanye dan 2 bulan setelah kampanye."

# Evaluasi hasil uji-t 1 bulan setelah kampanye
if p_value_1bulan < alpha:
    print("Kampanye berhasil setelah 1 bulan. H0 ditolak.")
    print(H1_1bulan)
else:
    print("Tidak cukup bukti untuk mengatakan bahwa kampanye berhasil setelah 1 bulan. H0 diterima.")
    print(H0_1bulan)

# Evaluasi hasil uji-t 2 bulan setelah kampanye
if p_value_2bulan < alpha:
    print("\n" "Kampanye berhasil setelah 2 bulan. H0 ditolak.")
    print(H1_2bulan)
else:
    print("Tidak cukup bukti untuk mengatakan bahwa kampanye berhasil setelah 2 bulan. H0 diterima.")
    print(H0_2bulan)

# Cetak nilai p_value dan t_statistic
print("\nNilai p-value (1 bulan pasca kampanye):", p_value_1bulan)
print("Nilai t-statistic (1 bulan pasca kampanye):", t_statistic_1bulan)
print("Nilai p-value (2 bulan pasca kampanye):", p_value_2bulan)
print("Nilai t-statistic (2 bulan pasca kampanye):", t_statistic_2bulan)
```

# **NO 3**

```python
# Pisahkan data untuk kelompok laki-laki dan perempuan 1 bulan setelah kampanye
male_1month_usage = df[df['sex'] == 0]['Post_usage_1month']
female_1month_usage = df[df['sex'] == 1]['Post_usage_1month']

# Pisahkan data untuk kelompok laki-laki dan perempuan 2 bulan setelah kampanye
male_2month_usage = df[df['sex'] == 0]['post_usage_2ndmonth']
female_2month_usage = df[df['sex'] == 1]['post_usage_2ndmonth']

alpha = 0.05  # Tingkat signifikansi

# Uji-t independen untuk 1 bulan setelah kampanye
# equal_var=False digunakan untuk mengasumsikan bahwa varians di kedua kelompok tidak sama.
# lebih aman dan umum digunakan karena tidak membuat asumsi yang kuat tentang kesamaan varians
# sehingga hasilnya lebih konservatif dan dapat digeneralisasi lebih baik.
t_statistic_1month, p_value_1month = stats.ttest_ind(male_1month_usage, female_1month_usage, equal_var=False)

# Uji-t independen untuk 2 bulan setelah kampanye
t_statistic_2month, p_value_2month = stats.ttest_ind(male_2month_usage, female_2month_usage, equal_var=False)

# Hipotesis
H0_1month = "Tidak ada perbedaan yang signifikan antara rata-rata penggunaan kartu kredit kelompok laki-laki dan perempuan setelah 1 bulan kampanye."
H1_1month = "Terdapat perbedaan yang signifikan antara rata-rata penggunaan kartu kredit kelompok laki-laki dan perempuan setelah 1 bulan kampanye."

H0_2month = "Tidak ada perbedaan yang signifikan antara rata-rata penggunaan kartu kredit kelompok laki-laki dan perempuan setelah 2 bulan kampanye."
H1_2month = "Terdapat perbedaan yang signifikan antara rata-rata penggunaan kartu kredit kelompok laki-laki dan perempuan setelah 2 bulan kampanye."

# Evaluasi hasil uji-t independen 1 bulan setelah kampanye
if p_value_1month < alpha:
    print("Ada perbedaan yang signifikan antara kelompok laki-laki dan perempuan setelah 1 bulan kampanye. H0 ditolak.")
    print(H1_1month)
else:
    print("Tidak cukup bukti untuk mengatakan ada perbedaan signifikan antara kelompok laki-laki dan perempuan setelah 1 bulan kampanye. H0 diterima.")
    print(H0_1month)

# Evaluasi hasil uji-t independen 2 bulan setelah kampanye
if p_value_2month < alpha:
    print("\n" "Ada perbedaan yang signifikan antara kelompok laki-laki dan perempuan setelah 2 bulan kampanye. H0 ditolak.")
    print(H1_2month)
else:
    print("Tidak cukup bukti untuk mengatakan ada perbedaan signifikan antara kelompok laki-laki dan perempuan setelah 2 bulan kampanye. H0 diterima.")
    print(H0_2month)

# Cetak nilai p-value dan t-statistic untuk 1 bulan setelah kampanye (kelompok laki-laki dan perempuan)
print("\nNilai p-value (1 bulan setelah kampanye untuk kelompok laki-laki dan perempuan):", p_value_1month)
print("Nilai t-statistic (1 bulan setelah kampanye untuk kelompok laki-laki dan perempuan):", t_statistic_1month)
# Cetak nilai p-value dan t-statistic untuk 2 bulan setelah kampanye (kelompok laki-laki dan perempuan)
print("Nilai p-value (2 bulan setelah kampanye untuk kelompok laki-laki dan perempuan):", p_value_2month)
print("Nilai t-statistic (2 bulan setelah kampanye untuk kelompok laki-laki dan perempuan):", t_statistic_2month)
```
