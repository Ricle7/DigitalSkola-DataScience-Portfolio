# **NO 1**



1.   [50 poin] Buatlah 5 visualisasi masing-masing untuk kolom attrition_flag customer_age, gender, education_level, dan income_category. (Wajib ada pie chart, donut chart, vertical bar chart, dan horizontal bar chart).



```python
# Buat subplot dengan 5 baris dan 1 kolom
fig, axes = plt.subplots(5, 1, figsize=(8, 15))

# Visualisasi 1: Pie Chart untuk attrition_flag
attrition_counts = df['Attrition_Flag'].value_counts()
axes[0].pie(attrition_counts, labels=attrition_counts.index, autopct='%1.2f%%')
axes[0].set_title('Pie Chart Attrition Flag')

# Visualisasi 2: Donut Chart untuk gender
gender_counts = df['Gender'].value_counts()
axes[1].pie(gender_counts, labels=gender_counts.index, autopct='%1.2f%%', wedgeprops={'width': 0.5})
axes[1].set_title('Donut Chart Gender')

# Visualisasi 3: Bar Chart (vertikal) untuk education_level
education_counts = df['Education_Level'].value_counts()
sns.barplot(x=education_counts.index, y=education_counts.values, ax=axes[2])
axes[2].set_title('Vertical Bar Chart Education Level')
# Menambahkan teks angka di tiap bar
for i, v in enumerate(education_counts.values):
    axes[2].text(i, v + 1, str(v), ha='center')

# Visualisasi 4: Bar Chart (horizontal) untuk income_category
income_counts = df['Income_Category'].value_counts()
sns.barplot(x=income_counts.values, y=income_counts.index, orient='h', ax=axes[3])
axes[3].set_title('Horizontal Bar Chart Income Category')
# Menambahkan teks angka di tiap bar
for i, v in enumerate(income_counts.values):
    axes[3].text(v + 1, i, str(v), va='center')

# Visualisasi 5: Histogram untuk customer_age
sns.histplot(df['Customer_Age'], bins=10, kde=True, color='blue', ax=axes[4])
axes[4].set_title('Age Distribution')
# Menambahkan teks angka di atas tiap bin histogram
for bar in axes[4].patches:
    height = bar.get_height()
    axes[4].annotate(f'{int(height)}', (bar.get_x() + bar.get_width() / 2, height), ha='center', va='bottom')

# Mengoptimalkan tata letak subplot
plt.tight_layout()

plt.show()
```

5. [10 poin] Buatlah 1 visualisasi menggunakan heatmap yang memadukan lebih dari satu kolom (bebas antar kolom yang mana saja)

```python
# Membuat dictionary untuk mengganti nama kolom
column_mapping = {
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1': 'Column_1',
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2': 'Column_2'
}
# Mengganti nama kolom karena kepanjangan
df_rename = df.rename(columns=column_mapping)

# Memilih kolom numerik untuk analisis korelasi
numeric_data = df_rename.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi antara Variabel Numerik', fontsize=16)
plt.show()
```
