# **Measures of Dispersion**

```python
# Range
data_max = data[["SepalWidthCm"]].max()
data_min = data[["SepalWidthCm"]].min()

data_max - data_min
```

```python
# Inter-Quantile Range
Q1 = data["SepalWidthCm"].quantile(q = 0.25)
Q3 = data["SepalWidthCm"].quantile(q = 0.75)

IQR = Q3 - Q1
IQR
```

```python
# Outlier
upper_inner_fence = Q3 + 1.5 * IQR
lower_inner_fence = Q1 - 1.5 * IQR

upper_outer_fence = Q3 + 3 * IQR
lower_outer_fence = Q1 - 3 * IQR

print("upper_inner_fence: ", upper_inner_fence)
print("lower_inner_fence: ", lower_inner_fence)
print("upper_outer_fence: ", upper_outer_fence)
print("lower_outer_fence: ", lower_outer_fence, "\n")
```

```python
print("Above upper outer fence: ")
data["SepalWidthCm"][data["SepalWidthCm"]
                     > upper_outer_fence]
```

```python
# pencilan minor
print("Above upper inner fence: ")
data["SepalWidthCm"][data["SepalWidthCm"]
                     > upper_inner_fence]
```

```python
# pencilan minor
print("Below lower inner fence: ")
data["SepalWidthCm"][data["SepalWidthCm"]
                     > lower_inner_fence]
```

```python
# pencilan mayor
print("Below lower outer fence: ")
data["SepalWidthCm"][data["SepalWidthCm"]
                     > lower_outer_fence]
```
