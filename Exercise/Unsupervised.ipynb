# Data Preprocessing

# Modeling

Use K-Means and Hierarchical Clustering

```python
# Feature Selection
df_fs = df.drop(columns='CustomerID')

# Label Encoding
df_encoded = df_fs.replace({'Gender': {'Female':0, 'Male':1}})

X = df_encoded
```

## K-Means Clustering

```python
# Elbow Method to identify optimum no. of clusters
inertia = []
range_of_clusters = list(range(2, 16))
for k in range_of_clusters:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

#Plotting Elbow Curve
plt.figure(figsize=(10, 4))
plt.plot(range_of_clusters, inertia)
plt.xticks(range_of_clusters)
plt.title('Elbow Curve')
plt.show()
```

Considering the above elbow curve, 4 number of clusters will be an optimum choice

```python
# K-Means Clustering with 4 clusters
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)

# Labeling the data
df_kmeans = df.copy()
df_kmeans['Cluster'] = kmeans.labels_
df_kmeans.head()
```

```python
# Categorical variable distribution among clusters
sns.countplot(data=df_kmeans, x='Cluster', hue='Gender')
plt.title('Gender Distribution Among Clusters')
plt.show()
```

```python
# Numerical variable distribution among clusters
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
cols = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
for i in range(len(cols)):
    sns.boxplot(data=df_kmeans, x='Cluster', y=cols[i], ax=axs[i])
fig.suptitle('Numerical Variable Distribution Among Clusters')
plt.tight_layout()
plt.show()
```

**Cluster Insights:**<br>
All the clusters have almost identical annual income distributions. So, let us analyze customer clusters based on gender, age, and spending score. <br>
Cluster 0: High age customers with low spending score and high annual income<br>
Cluster 1: High age customers with high spending score and high annual income<br>
Cluster 2: Low age customer with high spending score and medium annual income<br>
Cluster 3: Low age customer with high spending score and low annual income

## Hierarchical Clustering

There are two types of hierarchical clustering: Agglomerative and Divisive. We will use Agglomerative Hierarchical Clustering

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
X
```

```python
# Plotting dendrogram to identify optimum no. of clusters
mergings = linkage(X, method="ward")
plt.figure(figsize=(12, 6))
dendrogram(mergings)
plt.axhline(y=3, color='r', linestyle='--')
plt.title('Dendrogram')
plt.show()
```

```python
# Agglomerative clustering with 4 clusters
aggclust = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
aggclust.fit(X)

#Labeling the data
df_agg = df.copy()
df_agg['Cluster'] = aggclust.labels_
df_agg.head()
```

```python
# Categorical variable distribution among clusters
sns.countplot(data=df_agg, x='Cluster', hue='Gender')
plt.title('Gender Distribution Among Clusters')
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15,4))
cols = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
for i in range(len(cols)):
    sns.boxplot(data=df_agg, x='Cluster', y=cols[i], ax=axs[i])
fig.suptitle('Numerical Variable Distribution Among Clusters')
plt.tight_layout()
plt.show()
```
