```python
import pandas as pd
import numpy as np
from collections import Counter

# import ML Library
from sklearn import svm, preprocessing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier

# import evaluation metrics
from sklearn.metrics import classification_report, accuracy_score

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/Human%20Capital.csv')
data
```

```python
data['is_promoted'].value_counts()
```

```python
# # select desired columns

# data = data[["department", "education","gender", "recruitment channel",
#            "no_of_trainings","age","previous_year_rating","awards_won","avg_training_score","]]
```

```python
data_encoded = pd.get_dummies(data, prefix_sep="_")
data_encoded = data_encoded.dropna()
data_encoded.head()
```

```python
x = data_encoded.drop(columns="is_promoted")
y = data_encoded["is_promoted"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
print(Counter(y_train))
```

```python
clf = svm.SVC(C=1, kernel="rbf")
clf.fit(X_train, y_train)
```

```python
y_predict = clf.predict(X_test)

print("Accuracy :", accuracy_score(y_test, y_predict))

print(classification_report(y_test, y_predict))
```

```python
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1)
clf.fit(X_train, y_train)
```

```python
y_predict = clf.predict(X_test)

print("Accuracy :", accuracy_score(y_test, y_predict))

print(classification_report(y_test, y_predict))
```
