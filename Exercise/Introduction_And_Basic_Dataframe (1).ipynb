# **Membuat Dataframe dari 2D Array**

```python
my_2darray = np.array([[1,2,3],[4,5,6]])
print(my_2darray)
```

```python
df_array2 = pd.DataFrame(data=my_2darray,
                         index=range(0,2),
                         columns=[1,2,3])
print(df_array2)
```

```python
second_array = my_2darray[0:2,0:2]
second_array
```

```python
#Latihan buat dari my_2darray
df_2darray = pd.DataFrame(data=second_array,
                          index=range(0,2),
                          columns=['A','B'])
df_2darray
```

# Series with Dictionary

```python
my_series = pd.Series({"Jabar":"Bandung","Jateng":"Semarang","Jatim":"Surabaya","Bali":"Denpasar"})
my_series
```

```python
# Latihan buat series
# Gunakan Item
df_series = pd.DataFrame(data=my_series.items(),
                         index=range(0,4),
                         columns=['Provinsi', 'Ibu Kota'])
df_series
```

```python
#isi value per column
df = pd.DataFrame({"A": [1,2,3,4,5],
                   "B": [6,7,8,9,10],
                   "C": [11,12,13,14,15]})
df.head()
```

```python
#
df_series2 = pd.DataFrame({"Provinsi":["Jabar","Jateng","Jatim","Bali"],
                           "Ibu Kota":["Bandung","Semarang","Surabaya","Denpasar"]})
df_series2
```

# Selecting

```python
#Menggunakan iloc[] -> pada posisi di indeks, jadi akan mencari nilai DataFrame yang berada pada index x
print(df.iloc[2])
```

```python
#Menggunakan iloc[] -> ambil nilai pada indeks baris ke-o dan indeks kolom ke-0
print(df.iloc[0][0])
```

```python
#Menggunakan iloc[] -> ambil semua elemen indes 0-1
print(df.iloc[:,0:2])
```

```python
#Menggunakan at -> ambil index label x dari nama kolom y
print(df.at[2,'A'])
```

```python
#Menggunakan iat -> ambil index baris x dari nama kolom y
print(df.iat[4,2])
```
