## Cleaning 'Price ($)'

Cleaning the 'Price ($)' column by removing symbols and converting it to integer

```python
#cleaning
df['Price ($)'] = df['Price ($)'].str.replace('$','')
df['Price ($)'] = df['Price ($)'].str.replace(',','')
df['Price ($)'] = df['Price ($)'].str.replace(' ','')
df['Price ($)'] = df['Price ($)'].astype(int)
# Menampilkan ringkasan statistik deskriptif setelah membersihkan kolom 'Price ($)'
df.describe()
```

## Grouping and Aggregating Data

Grouping and calculating averages for high-level analysis

## Data Categorization and Binning

Creating price categories using quantiles

```python
# mengkategorisasikan nilai Price menjadi 4 kategori dengan jumlah ukuran yang sama, misal Bronze, Silver, Gold, Platinum
bin_labels = ['Bronze', 'Silver', 'Gold', 'Platinum']
results, bin_edges = pd.qcut(df['Price ($)'], q = 4, labels = bin_labels, retbins = True)
```

```python
results
```

```python
bin_edges
```

```python
results_table = pd.DataFrame(zip(bin_edges, bin_labels), columns =['Threshold', 'Tier'])
results_table
```

```python
# Menambahkan kolom 'results' ke DataFrame
df['results'] = results
df
```

```python
# Menunjukkan distribusi
df['results'].value_counts()
```

```python
# Mengatur batas kategori harga secara manual
bins = [0, 1000, 2000, 3000, 4000, 5000, 10000]
labels = [1, 2, 3, 4, 5, 6]
df['binned_battery'] = pd.cut(df['Battery Capacity (mAh)'], bins = bins, labels = labels)
```

```python
df
```

```python
# Menunjukkan distribusi kategori kapasitas baterai
df['binned_battery'].value_counts()
```
